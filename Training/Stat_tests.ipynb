```python
import numpy as np
import pandas as pd
import pylab
import scipy.stats as stats
from tqdm.notebook import tqdm
```

```python
def experiment():
    cards = np.random.binomial(1, 0.5, size=10)
    guess = np.random.binomial(1, 0.5, size=10)
    result.append((np.sum(cards == guess) >= 9))
    return result
```

```python
n = 1000
result = []

np.mean([any([experiment() for _ in range(n)]) for _ in tqdm(range(1))])
```

```python
stats.probplot(pd.Series(result), dist="uniform", plot=pylab)
pylab.show()
```
