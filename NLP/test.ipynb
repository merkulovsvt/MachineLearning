```python
import numpy as np
import seaborn as sns


def func(x):
    return 10 * x ** 3 + 6 * x ** 2 - 12 * x + 5 * np.sin(x) * np.cos(x)


def func_der(x):
    return 30 * x ** 2 + 12 * x - 12 + 6 * np.cos(x)


X = np.linspace(-1, 1, 30)
Y = func(X)

sns.lineplot(x=X, y=Y)
```

```python
import numpy as np
import seaborn as sns


def pred_func(b, k1, k2, k3, x):
    return b + k1 * x + k2 * x ** 2 + k3 * x ** 3


lr = 0.01
b, k1, k2, k3 = 1, 1, 1, 1

for _ in range(10000):
    y_pred = pred_func(b, k1, k2, k3, X)

    b -= 2 * np.mean((y_pred - Y)) * lr
    k1 -= 2 * np.mean((y_pred - Y) * X) * lr
    k2 -= 2 * np.mean((y_pred - Y) * X ** 2) * lr
    k3 -= 2 * np.mean((y_pred - Y) * X ** 3) * lr

    print(b, k1, k2, k3)

sns.lineplot(x=X, y=Y, zorder=1)
sns.scatterplot(x=X, y=pred_func(b, k1, k2, k3, X), zorder=2)
```
