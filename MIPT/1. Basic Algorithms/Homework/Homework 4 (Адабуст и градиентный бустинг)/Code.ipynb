CatBoost

```python
y = data['Attrition_Flag']
X = data.drop(['Attrition_Flag', 'CLIENTNUM'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
model = CatBoostClassifier()
param_grid = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'iterations': [50, 100, 200]
}

grid_search_result = model.grid_search(param_grid=param_grid, X=X, y=y, plot=True, cv=6)
```

```python
print(classification_report(grid_search_result.predict(X_test), y_test))
```

```python
zaza= CatBoostClassifier(grid_search_result)
zaza.fit(X_test, y_test)
```
